```python
import pandas as pd
```

```python
results = pd.read_csv('C:\\Users\\Checkout\\Desktop\\Pet Project\\F1\\Ergast Data\\Cleaned Data\\results.csv')
races = pd.read_csv('C:\\Users\\Checkout\\Desktop\\Pet Project\\F1\\Ergast Data\\Cleaned Data\\races.csv')
quali = pd.read_csv('C:\\Users\\Checkout\\Desktop\\Pet Project\\F1\\Ergast Data\\Cleaned Data\\qualifying.csv')
drivers = pd.read_csv('C:\\Users\\Checkout\\Desktop\\Pet Project\\F1\\Ergast Data\\Cleaned Data\\drivers.csv')
constructors = pd.read_csv('C:\\Users\\Checkout\\Desktop\\Pet Project\\F1\\Ergast Data\\Cleaned Data\\constructors.csv')
circuit = pd.read_csv('C:\\Users\\Checkout\\Desktop\\Pet Project\\F1\\Ergast Data\\Cleaned Data\\circuits.csv')
status = pd.read_csv('C:\\Users\\Checkout\\Desktop\\Pet Project\\F1\\Ergast Data\\Cleaned Data\\status.csv')
```

```python
df1 = pd.merge(races,results,how='inner',on=['raceId'])
df2 = pd.merge(df1,quali,how='inner',on=['raceId','driverId','constructorId'])
df3 = pd.merge(df2,drivers,how='inner',on=['driverId'])
df4 = pd.merge(df3,constructors,how='inner',on=['constructorId'])
df5 = pd.merge(df4,circuit,how='inner',on=['circuitId'])
df6 = pd.merge(df5,status,how='inner',on=['statusId'])
```

```python
data = df6.drop(['constructorId', 'circuitId', 'location', 'driverRef', 'position_x', 'laps', 'round', 'resultId', 'positionText', 'rank', 'qualifyId', 'fastestLapSpeed', 'number_x', 'fastestLap', 'number_y', 'fastestLapTime', 'name_x', 'q2', 'q1', 'milliseconds', 'code', 'q3', 'statusId'], axis = 1)
```

```python
data.rename(columns={'year': 'Year', 'positionOrder': 'Final Position', 'points': 'Points', 'position_y': 'Quali Position', 'forename': 'Forename', 'surname': 'Surname', 'nationality_x': 'Nationality', 'name_y': 'Constructor', 'nationality_y': 'Base', 'name': 'Name', 'country': 'Location', 'status': 'Status'}, inplace=True)
```

```python
data.to_csv("C:\\Users\\Checkout\\Desktop\\Pet Project\\F1\\Ergast Data\\data_main.csv", index=False)
```
